```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Define the image path
image_path = "pexels-jhawley-57905.jpg"

# Read the image in grayscale
image_details = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Check if the image was successfully read
if image_details is not None:
    print("Image loaded successfully")

    # Calculate histogram for the grayscale image
    name_for_histogram = cv2.calcHist([image_details], [0], None, [256], [0, 256])

    # Flatten the histogram
    name_for_histogram = name_for_histogram.flatten()

    # Plot the image and histogram side by side
    plt.figure(figsize=(12, 6))

    # Plot the image
    plt.subplot(1, 2, 1)
    plt.imshow(image_details, cmap='gray')
    plt.title('Grayscale Image')
    plt.axis('off')  # Hide the axis

    # Plot the histogram
    plt.subplot(1, 2, 2)
    plt.plot(name_for_histogram)
    plt.title('Histogram')
    plt.xlabel('Pixel value')
    plt.ylabel('Frequency')
    plt.grid(True)

    plt.show()
else:
    print("Error loading image")
```
